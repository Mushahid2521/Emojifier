```

def sentences_to_indices(X, word_to_index, max_len):
    
    m = X.shape[0]                                

    X_indices = np.zeros((m,max_len))
    
    for i in range(m):                              
        
        sentence_words = X[i].lower().split()
        
        j = 0


        for w in sentence_words:
            X_indices[i, j] = word_to_index[w]
            j = j+1
            
    
    return X_indices
```

```
from keras.models import load_model
import pickle

def emojifier(input_text):
  emo_model = load_model('trained_emojify.h5')
  
  with open('word2index.pickle', 'rb') as handle:
    word_to_index = pickle.load(handle)
     
  x = np.array([input_text])
  x_indices = sentences_to_indices(x, word_to_index, maxLen)
  return input_text +' '+  label_to_emoji(np.argmax(emo_model.predict(x_indices)))
  
```
